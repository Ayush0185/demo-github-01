Arithmetic operators

```python
num_1 = 20
num_2 = 10

#addition operators
sum = num_1 + num_2
print(sum)

#subtraction

diff = num_1 - num_2
print(diff)

# product

pro = num_1 * num_2
print(pro)

#division

quo = num_1/ num_2
print(quo)

#exponent

exp = num_1 **num_2  #20^10
print('exponent =',exp)


#modulus
mod = num_1 % num_2
print('reminder=',mod)
```

Assignment Operators

```python
a = 5
print(a)
```

```python
a = 5
a += 5 # a = a  +  5
print(a)
```

```python
b = 5
b -= 2 # b = b - 2
print(b)
```

Comparison Operators

```python
a = 5
b = 10

print(a == b) #equal to
print(a != b)#not equal to
print(a > b)#greater than symbol
print(a < b)#less than
```
